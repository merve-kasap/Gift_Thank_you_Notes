```python
list = list[['Gift Giver', 'Gift Giver Address', 'Gift Message', 'Name']]
```

```python
list['Gift Message'] = list['Gift Message'].fillna('Thanks')
```

```python
list.dropna(subset=['Gift Giver Address'])
```

```python
def create_thank_you_messages(df):
    thank_you_messages = []
    
    grouped = df.groupby('Gift Giver').agg({'Name': ', '.join, 'Gift Message': ' '.join, 'Gift Giver Address': 'first'}).reset_index()

    for index, row in grouped.iterrows():
        message = f"Dear {row['Gift Giver']},\n\nWe just received your gifts, the '{row['Name']}'. Thank you so much! Your messages, '{row['Gift Message']}', truly touched our hearts. We appreciate your thoughtfulness and kindness.\n\n"
        if pd.notnull(row['Gift Giver Address']):
            message += f"Address: {row['Gift Giver Address']}\n\n"
        message += "Best,\nMerve & Ben"
        thank_you_messages.append(message)

    return thank_you_messages
```

```python
df = list  # your DataFrame
thank_you_messages = create_thank_you_messages(df)

# print the thank you messages
for msg in thank_you_messages:
    print(msg)
    print("\n-----------------\n")
```
